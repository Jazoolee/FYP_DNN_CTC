```python
target_pos = []
for i in range(-160, 160, 5):
    for j in range (-225, 45, 5):
        for k in range(-45, 225, 5):
            # convert the degree to radiant position
            i_rad = i * 3.141592653589793 / 180.0
            j_rad = j * 3.141592653589793 / 180.0
            k_rad = k * 3.141592653589793 / 180.0
            target_pos.append([i_rad, j_rad, k_rad])
    pass
print(len(target_pos))
```

```python
import numpy as np
import csv
```

```python
randomized_pos = np.random.permutation(target_pos)
```

```python
len(randomized_pos)
for i in range(100):
    print(randomized_pos[i])
```

```python
# save to csv
with open('target_positions.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['pos1', 'pos2', 'pos3'])
    for i in range(100):
        writer.writerow(randomized_pos[i])
```
